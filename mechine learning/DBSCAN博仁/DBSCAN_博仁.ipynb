```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets#使用scikit-learn內建的資料
%matplotlib inline
# %matplotlib inline可以在Ipython編譯器裡直接使用，功能是可以內嵌繪圖，並且可以省略掉plt.show()這一步。
X1, y1=datasets.make_circles(n_samples=5000, factor=.6,
                                      noise=.05)
'''
n_samples：int，可選（默認= 100）生成的總點數。 如果為奇數，則內圓將比外圓多一點。
factor：0 <double <1（默認= .8）內圓和外圓之間的比例因子。
noise：兩倍或無（默認=無）加到數據中的高斯噪聲的標準偏差。             (用途??
shuffle : bool布林值, optional預設true (default=True)Whether to shuffle the samples是否洗牌。.
Returns:	
X : array of shape [n_samples, 2]                 (為什麼x,y?各意義?
The generated samples.

y : array of shape [n_samples]
The integer labels (0 or 1) for class membership of each sample.
https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html

'''
X2, y2 = datasets.make_blobs(n_samples=1000, n_features=10, centers=[[1.2,1.2]], cluster_std=[[.1]],
               random_state=9)#生成聚類算法的測試數據 
'''
待生成的樣本的總數=1000
是每個樣本的特徵數=10                                     (wt mean 10未知數?
centers=要生成的樣本中心（類別）數，或者是確定的中心點       (中心點為1.2,1.2??
cluster_std表示每個類別的方差                               ([[.1]]??
#random_state=預設為NONE,若設為INT則random_state是隨機數生成器使用的種子
'''
X = np.concatenate((X1, X2))#數組拼接
#https://blog.csdn.net/brucewong0516/article/details/79158758
#https://blog.csdn.net/zyl1042635242/article/details/43162031
plt.scatter(X[:, 0], X[:, 1], marker='o')#繪製散點圖 ,X [:,0]就是取所有行的第0個數據, marker:散點的形狀參數,預設為o為圈
plt.show()
```

```python
from sklearn.cluster import DBSCAN
y_pred = DBSCAN(eps=0.1,metric='manhattan').fit_predict(X)
                
'''
eps： DBSCAN算法參數,默認值是0.5,即我們的 -鄰域的距離閾值，和样本距離超過的樣本點不在 -鄰域內。一般需要通過在多組值裡面選擇一個合適的閾值。
eps過大，則更多的點會落在核心對象的 -鄰域，此時我們的類別數可能會減少，本來不應該是一類的樣本也會被劃為一類。反之則類別數可能會增大，本來是一類的樣本卻被劃分開。

min_samples： DBSCAN算法參數，即樣本點要成為核心對象所需要的 -鄰域的樣本數閾值。默認值是5.
一般需要通過在多組值裡面選擇一個合適的閾值。通常和eps一起調參。
在eps一定的情況下，min_samples過大，則核心對象會過少，此時簇內部分本來是一類的樣本可能會被標為噪音點，類別數也會變多。反之min_samples過小的話，則會產生大量的核心對象，可能會導致類別數過少。
min_samples定義:一個點附近的樣品數量（或總重量）被視為核心點。 這包括點本身(如果該點為樣本點)。
 
metric：最近鄰距離度量參數。可以使用的距離度量較多，一般來說DBSCAN使用默認的歐式距離（即p=2的閔可夫斯基距離） 





'''
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)#讓KMeans分成3類
kmeans.fit(X)#開始訓練
y_kmeans = kmeans.predict(X)#應用訓練模組在數據上

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')#x座標為x所有的第0個資料,y座標為x所有的第1個資料,c根據各類變色,
#cmap:顏色圖實例或註冊的顏色圖名稱。 僅當c是浮點數數組時才使用cmap,viridis	藍-綠-黃
#cmap:https://www.cnblogs.com/denny402/p/5122594.html
centers = kmeans.cluster_centers_#找各類中心                    (沒有宣告成陣列怎麼是陣列???
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
from sklearn.cluster import DBSCAN
y_pred = DBSCAN(eps=0.1,metric='euclidean').fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```
