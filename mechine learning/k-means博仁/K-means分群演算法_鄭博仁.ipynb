```python
%matplotlib inline
# %matplotlib inline可以在Ipython編譯器裡直接使用，功能是可以內嵌繪圖，並且可以省略掉plt.show()這一步。
import matplotlib.pyplot as plt#繪圖函示庫
import seaborn as sns; sns.set()#Seaborn 套件是以 matplotlib 為基礎建構的高階繪圖套件，讓使用者更加輕鬆地建立圖表
import numpy as np#Numpy 是 Python的一個重要模組,透過整合其他低階語言同時擁有效能和高效率的開發,主要用於資料處理上
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)#生成聚類算法的測試數據 
#n_samples=待生成的樣本的總數,centers=要生成的樣本中心（類別）數，或者是確定的中心點,
#cluster_std=cluster_std表示每個類別的方差，例如我們希望生成2類數據，其中一類比另一類具有更大的方差，可以將cluster_std設置為[1.0,3.0]。
#random_state=預設為NONE,若設為INT則random_state是隨機數生成器使用的種子；
plt.scatter(X[:, 0], X[:, 1], s=50);#繪製散點圖 ,X [:,0]就是取所有行的第0個數據,s=點大小 預設為20
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import DBSCAN
y_pred = DBSCAN(eps=0.1,metric='manhattan').fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)#讓KMeans分成4類
kmeans.fit(X)#開始訓練
y_kmeans = kmeans.predict(X)#應用訓練模組在數據上

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')#x座標為x所有的第0個資料,y座標為x所有的第一個資料,c根據維度變色(?),
#cmap:顏色圖實例或註冊的顏色圖名稱。 僅當c是浮點數數組時才使用cmap

centers = kmeans.cluster_centers_#挑出4個中心數據聚類中心
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);#x座標為center[]的第一個數據,y座標同理,黑色,透明度0.5
```
