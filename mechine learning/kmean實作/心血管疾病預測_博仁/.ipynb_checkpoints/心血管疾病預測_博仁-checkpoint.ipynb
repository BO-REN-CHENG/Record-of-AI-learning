```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style("whitegrid") # 設置畫圖的風格
sns.set_context("paper") 
df = pd.read_csv("cardio_train.csv",';') #讀取CSV檔
df.drop(['id'],axis=1,inplace=True)
'''
https://blog.csdn.net/b876144622/article/details/80781917
https://www.itread01.com/content/1548805503.html
'''
```

```python
df.tail()#印出最後5個數據
#Python 資料處理筆記 - 使用Panda進行數據處理
#https://medium.com/@yanweiliu/python-pandas%E4%BD%BF%E7%94%A8%E7%AD%86%E8%A8%98-a4682e254d90
```

```python
## dropna可去掉包含nan的行或列
df.dropna(
    axis=0,     # 0: 对列作; 1: 对行操作
    how='any',   # 'any': 只要存在 NaN 就 drop 掉; 'all': 必须全部是 NaN 才 drop 
    inplace=True)
df = df.reset_index(drop=True)#reset_index可以還原索引，重新變為默認的整型索引 ,drop為False則索引列會被還原為普通列，否則會丟失

#pandas中關於set_index和reset_index的用法
#https://blog.csdn.net/jingyi130705008/article/details/78162758
```

```python
df.tail()#印出最後5個數據確認有沒有東西被刪掉
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = df.drop(['cardio'],axis=1)#x=刪掉的cardio行????????
y = df['cardio']#y重創cardio行????
x_train, x_test, y_train, y_test = train_test_split(X, y,#取總資料的80%做訓練集,20%做測試集
                                                    #xtrain,ytrain是訓練集又各自切一半嗎??????????
                                                    test_size=0.2)
'''
train_test_split(train_data,train_target,test_size=0.4, random_state=0)
train_data ：所要劃分的樣本特徵集

train_target ：所要劃分的樣本結果

test_size ：樣本佔比，如果是整數的話就是樣本的數量

random_state ：是隨機數的種子。

隨機數種子：其實就是該組隨機數的編號，在需要重複試驗的時候，保證得到一組一樣的隨機數。比如你每次都填1 ，其他參數一樣的情況下你得到的隨機數組是一樣的。但填0 或不填，每次都會不一樣。

隨機數的產生取決於種子，隨機數和種子之間的關係遵從以下兩個規則：

種子不同，產生不同的隨機數；種子相同，即使實例不同也產生相同的隨機數。
#擔心overfitting的情況發生的話
#進行機器學習的時候，就需區分 訓練用data(總資料的80%) 以及 測試用data(總資料的20%)，
#函數train_test_split可以幫我們自動做到，他會自動幫我們亂數取出總資料的20%，來當成測試的時候使用
#這些 測試用data 將不會參與訓練的過程！

'''
#[Python]初心者筆記11(線性回歸區分訓練資料以及測試資料,train and test data, 使用套件直接獲得現實世界數據)
#https://dotblogs.com.tw/kevinya/2018/06/11/093633
#Sklearn-train_test_split隨機劃分訓練集和測試集
#https://blog.csdn.net/CherDW/article/details/54881167
#機器學習sklearn中的train_test_split()函數
#https://blog.51cto.com/12831900/2300061
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(x_train)#用x_train數據訓練模駔????????????
```

```python
y_pred = kmeans.predict(x_test)#把測試數據應用在模組中 結果存在y_pred??
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_pred)
#混淆矩陣及confusion_matrix函數的使用
#https://blog.csdn.net/m0_38061927/article/details/77198990
sns.set(font_scale=1.5)#格子內數字字型大小
#seaborn.set
#https://seaborn.pydata.org/generated/seaborn.set.html
'''
Parameters:	
context : string or dict

Plotting context parameters, see plotting_context()

style : string or dict

Axes style parameters, see axes_style()

palette : string or sequence

Color palette, see color_palette()

font : string

Font family, see matplotlib font manager.

font_scale : float, optional

Separate scaling factor to independently scale the size of the font elements.

color_codes : bool

If True and palette is a seaborn palette, remap the shorthand color codes (e.g. “b”, “g”, “r”, etc.) to the colors from this palette.

rc : dict or None

Dictionary of rc parameter mappings to override the above.
'''
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
#square : 为‘True’时，整个网格为一个正方形 
#annot : 如果为True，则将数据值写入每个单元格中
# fmt='d':顯示型態為decimal 十進位制
#cbar : 是否绘制颜色条：colorbar，默认绘制(依照數據大小做出的顏色漸層條)
                                                    #為什麼mat要.t????????
#python3.x-seaborn.heatmap随笔
#https://zhuanlan.zhihu.com/p/35494575
plt.xticks(fontsize=20)#x軸字體大小
plt.yticks(fontsize=20)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
