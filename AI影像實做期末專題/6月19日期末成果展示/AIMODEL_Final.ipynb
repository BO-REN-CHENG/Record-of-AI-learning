```python
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
data = []#建立data label陣列
label = []
for i in os.listdir("me/"):#for迴圈簡化讀取過程
    img = cv2.imread('me/'+i)
#     img = cv2.imread('C:/absUsers/z1999/OneDrive/桌面/train/me/'+i)
    img = cv2.resize(img,(200,200),interpolation=cv2.INTER_CUBIC)#統一resize成相同大小200*200
    
    data.append(img)
    label.append(0)#分類為是我的統一統一label為0
    
    
for i in os.listdir("not me/"):#for迴圈簡化讀取過程
    img = cv2.imread('not me/'+i)
    print(i)
    img = cv2.resize(img,(200,200),interpolation=cv2.INTER_CUBIC)#統一resize成相同大小200*200
    
    data.append(img)
    label.append(1)#分類為不是我的統一label成1
# cv2.imshow('img',img)
cv2.waitKey(0)
```

```python
data = np.array(data)
label = np.array(label)
```

```python
data.shape#2247張 長寬200*200 3為rgb
```

```python
data_reshape = np.reshape(data,(2247,200*200*3))#統一攤平,機器學習需要攤平
data_reshape.shape
```

```python
plt.imshow(data_reshape)#觀察攤平結果
data_reshape
```

```python
from sklearn.model_selection import train_test_split#設定隨機森林參數
x_train, x_test, y_train, y_test = train_test_split(data_reshape, label,
                                                    test_size=0.2)
```

```python
# from sklearn.cluster import KMeans 
from sklearn.ensemble import RandomForestClassifier#建立隨機森林模型
# kmeans = KMeans(n_clusters=2)
# kmeans.fit(x_train)

clf = RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
# y_pred = kmeans.predict(x_test)
# y_pred
```

```python
y_pred = clf.predict(x_test)#觀察預測結果
y_pred
```

```python
from sklearn.metrics import confusion_matrix      #使用混淆矩陣觀察預測的好不好
from sklearn.metrics import classification_report  
import seaborn as sns 
mat = confusion_matrix(y_test, y_pred)
#plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('true label')
plt.xlabel('predicted label');
target_names = ['我','不是我']
print(classification_report(y_test,y_pred,target_names=target_names))
```

# 開啟鏡頭開始人臉辨識是不是我

```python
# 導入抓取人臉分類器
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# 從視訊鏡頭擷取影片
cap = cv2.VideoCapture(0)


while True:

    # Read the frame
    
    _, img = cap.read()
# 轉成灰階
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 偵測臉部
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
# 繪製人臉部份的方框
#     cv2.waitKey(30)
    for (x, y, w, h) in faces:
#             cv2.waitKey(100)
        crop_img = img[y:y+h, x:x+w]
        img_resize = cv2.resize(crop_img,(200,200),interpolation=cv2.INTER_CUBIC)
        img_flatten = np.reshape(img_resize,(1,200*200*3))
        y_pred = clf.predict(img_flatten)
        
        if y_pred == 0:
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            cv2.putText(img,'me' ,(x, y),cv2.FONT_HERSHEY_COMPLEX,1,(0, 255, 0),1)
#             cv2.waitKey(100)

        else:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 0), 2)
            cv2.putText(img,'not me' ,(x, y),cv2.FONT_HERSHEY_COMPLEX,1,(255, 255, 0),1)
#             cv2.waitKey(100)

        #cv2.imwrite('crop.jpg', crop_img)
        #cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
# 顯示成果
    cv2.namedWindow('img', cv2.WINDOW_NORMAL)  #正常視窗大小
    cv2.imshow('img', img)                     #秀出圖片
# Stop if escape key is pressed
    k = cv2.waitKey(30) & 0xff
    if k==27:
        break
        
# Release the VideoCapture object
cap.release()
cv2.destroyAllWindows()
```
